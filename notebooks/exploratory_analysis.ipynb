```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('../data/processed/superstore_clean.csv')
```

```python
base.head()
```

```python
base.info()
```

```python
base.describe()
```

```python
# Criando coluna de ano para análise temporal

base['Order Date'] = pd.to_datetime(base['Order Date'])

base['Year'] = base['Order Date'].dt.year

base.head()
```

```python
# Vendas e lucro por ano
sales_year = base.groupby('Year')[['Sales', 'Profit']].sum()
sales_year.plot(kind='bar', figsize=(10, 6))
plt.title('Vendas e Lucro por Ano')
plt.ylabel('Valor')
plt.show()
```

### Insight

Observa-se crescimento nas vendas ao longo dos anos, indicando expansão do negócio. 
O lucro acompanha essa tendência, mostrando que o aumento de vendas não comprometeu a rentabilidade.

```python
# Produto que mais vendeu
top_products = base.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 Produtos por Vendas')
plt.ylabel('Vendas')
plt.show()
```

```python
# Categorias que mais geraram receita
category_sales = base.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales.plot(kind='bar', figsize=(8, 5))
plt.title('Vendas por Categoria')
plt.ylabel('Vendas')
plt.show()
```

### Insight

A categoria Technology apresenta maior volume de vendas, indicando maior demanda ou maior ticket médio em comparação às demais categorias.


```python
# Regiao com mais vendas
region_sales = base.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar', figsize=(8, 5))
plt.title('Vendas por Região')
plt.ylabel('Vendas')
plt.show()
```

### Insight

A região com maior volume de vendas demonstra maior concentração de clientes ou maior presença da empresa nessa área.

```python
# Qual segmento de clientes compra mais
segment_sales = base.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
segment_sales.plot(kind='bar', figsize=(8, 5))
plt.title('Vendas por Segmento')
plt.ylabel('Vendas')
plt.show()
```

```python
# Qual tempo medio de entrega dos pedidos
base['Ship Date']= pd.to_datetime(base['Ship Date'])
base['Delivery Time'] = (base['Ship Date'] - base['Order Date']).dt.days
plt.figure(figsize=(10, 5))
sns.histplot(base['Delivery Time'], bins=30, kde=True)
plt.title('Tempo de Entrega dos Pedidos')
plt.xlabel('Dias')
plt.ylabel('Frequência')
plt.show()
```

### Insight

A maioria dos pedidos é enviada entre 3 e 6 dias, com média próxima de 4 dias, indicando um prazo logístico relativamente consistente.

## Conclusão Geral da Análise

A análise exploratória do dataset Superstore permitiu identificar padrões relevantes no desempenho de vendas, lucratividade e operação logística da empresa.

Observou-se que as vendas apresentam crescimento ao longo dos anos, indicando expansão do negócio e aumento da demanda. Além disso, algumas categorias e produtos se destacam por gerar maior volume de receita e lucro, sugerindo áreas estratégicas para investimento comercial.

A análise por região mostrou diferenças no volume de vendas entre mercados, o que pode refletir a distribuição de clientes ou a presença da empresa em determinadas localidades. Da mesma forma, os segmentos de clientes apresentam comportamentos distintos de compra, o que pode orientar estratégias de marketing e relacionamento.

Por fim, a avaliação do tempo de entrega revelou que a maioria dos pedidos é enviada em aproximadamente 3 a 6 dias, indicando um processo logístico relativamente consistente, embora existam alguns casos isolados de prazos maiores.

De forma geral, os resultados desta análise fornecem uma visão inicial do desempenho do negócio e servem como base para análises mais aprofundadas, como segmentação de clientes, avaliação de rentabilidade por produto e construção de dashboards para acompanhamento estratégico.
